# Importing libraries required

```python
import pandas as pd
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import plotly.graph_objects as go
```

```python
import plotly.figure_factory as ff
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(color_codes=True) #seaborn theme colour variations 
```

```python
#Loading data into data frame
us=pd.read_csv('unemployment data in usa.csv')
```

```python
# Displaying the first 5 rows of the dataset
us.head(5)
```

```python
#Checking for the number of rows and columns 
us.shape
```

```python
# Droping irrelevant colunms from the dataset
us= us.drop (['Date'],axis=1)
```

```python
us.head()
```

```python
#count the number of rows
us.count()
```

# Checking and Dropping missing or null values

```python
print (us.isnull ().sum())
```

```python
us = us.dropna()
us.count()
```

```python
print (us.isnull ().sum())
```

# Using charts to create visuals

```python
sns.boxplot (x=us['Primary_School'])
```

```python
sns.catplot(x="Year", y="High_School", hue="Year",
            kind="violin", bw=.25, cut=10,
            data=us)
```

```python
#Total number of death caused by covid across regions
#bar chart
sns.barplot(x = 'Year', y = 'Associates_Degree', data = us,
            label = 'WHO Region by Death cases', color = 'b', edgecolor = 'w')
plt.figure(figsize=(16, 75))
sns.despine(left = True, bottom = True)
plt.show()
```

```python
#Total number of death caused by covid across regions
#bar chart
sns.barplot(x = 'Year', y = 'Professional_Degree', data = us,
            label = 'WHO Region by Death cases', color = 'b', edgecolor = 'w')
plt.figure(figsize=(16, 75))
sns.despine(left = True, bottom = True)
plt.show()
```

```python
sns.catplot(x="Year", y="Men", kind="box", data=us)
```

```python
sns.catplot(x="Year", y="Women", hue="Year",
            kind="box", dodge=False, data=us)
```

```python
sns.catplot(x="Year", y="White", hue="Year", kind="swarm", data=us)
```

```python
sns.violinplot(x=us.Year, y=us.White)
```

```python
sns.violinplot(x=us.Year, y=us.Black)
```

```python
sns.catplot(x="Year", y="Asian", hue="Year",
            #kind="violin", inner="stick", split=True,
            palette="pastel", data=us)
```

```python
sns.catplot(x="Year", y="Hispanic", hue="Year",
            kind="violin", bw=.25, cut=10,
            data=us)
```

```python
plt.figure(figsize=(10,5))
c= us.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
labels = ['Primary_School', 'High_School', 'Associates_Degree', 'Professional_Degree', 'White', 'Black', 'Asian', 'Hispanic', 'Men', 'Women']
sizes = [us.loc[0,'Primary_School'],
         us.loc[0,'High_School'],
         us.loc[0,'Associates_Degree'],
         us.loc[0,'Professional_Degree'],
         us.loc[0,'White'],
         us.loc[0,'Black'],
         us.loc[0,'Asian'],
         us.loc[0,'Hispanic'],
         us.loc[0,'Men'],
         us.loc[0,'Women']
        ]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.show()
```
